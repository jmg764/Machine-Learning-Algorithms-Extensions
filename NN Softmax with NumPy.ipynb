## 1. MNIST

### Homework implementation

```python
digits=load_digits()
X = digits.data
y = digits.target

# Scale training features
X_scale = StandardScaler()
X = X_scale.fit_transform(digits.data)
```

```python
# Split the data into training and test set.  60% training and %40 test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
def convert_y_to_vect(y):
    y_vect = np.zeros((len(y), 10))
    for i in range(len(y)):
        y_vect[i, y[i]] = 1
    return y_vect
```

```python
# Convert digits to vectors
y_v_train = convert_y_to_vect(y_train)
y_v_test = convert_y_to_vect(y_test)
```

```python
# Sigmoid activation function

def f(z):
    return 1 / (1 + np.exp(-z))


def f_deriv(z):
    return f(z) * (1 - f(z))
```

```python
def setup_and_init_weights(nn_structure):
    W = {} #creating a dictionary i.e. a set of key: value pairs
    b = {}
    for l in range(1, len(nn_structure)):
        W[l] = r.random_sample((nn_structure[l], nn_structure[l-1])) #Return “continuous uniform” random floats in the half-open interval [0.0, 1.0). 
        b[l] = r.random_sample((nn_structure[l],))
    return W, b
```

```python
def init_tri_values(nn_structure):
    tri_W = {}
    tri_b = {}
    for l in range(1, len(nn_structure)):
        tri_W[l] = np.zeros((nn_structure[l], nn_structure[l-1]))
        tri_b[l] = np.zeros((nn_structure[l],))
    return tri_W, tri_b
```

```python
def feed_forward(x, W, b):
    a = {1: x} # create a dictionary for holding the a values for all levels
    z = { } # create a dictionary for holding the z values for all the layers
    for l in range(1, len(W) + 1): # for each layer
        node_in = a[l]
        z[l+1] = W[l].dot(node_in) + b[l]  # z^(l+1) = W^(l)*a^(l) + b^(l)
        a[l+1] = f(z[l+1]) # a^(l+1) = f(z^(l+1))
    return a, z
```

```python
def calculate_out_layer_delta(y, a_out, z_out):
    # delta^(nl) = -(y_i - a_i^(nl)) * f'(z_i^(nl))
    return -(y-a_out) * f_deriv(z_out) 


def calculate_hidden_delta(delta_plus_1, w_l, z_l):
    # delta^(l) = (transpose(W^(l)) * delta^(l+1)) * f'(z^(l))
    return np.dot(np.transpose(w_l), delta_plus_1) * f_deriv(z_l)
```

```python
# Back propagation algorithm

def train_nn(nn_structure, X, y, iter_num=3000, alpha=0.25):
        W, b = setup_and_init_weights(nn_structure)
        cnt = 0
        N = len(y)
        avg_cost_func = []
        print('Starting gradient descent for {} iterations'.format(iter_num))
        while cnt < iter_num:
            if cnt%1000 == 0:
                print('Iteration {} of {}'.format(cnt, iter_num))
            tri_W, tri_b = init_tri_values(nn_structure)
            avg_cost = 0
            for i in range(N):
                delta = {}
                # perform the feed forward pass and return the stored a and z values, to be used in the
                # gradient descent step
                a, z = feed_forward(X[i, :], W, b)
                # loop from nl-1 to 1 backpropagating the errors
                for l in range(len(nn_structure), 0, -1):
                    if l == len(nn_structure):
                        delta[l] = calculate_out_layer_delta(y[i,:], a[l], z[l])
                        avg_cost += np.linalg.norm((y[i,:]-a[l]))
                    else:
                        if l > 1:
                            delta[l] = calculate_hidden_delta(delta[l+1], W[l], z[l])
                        # triW^(l) = triW^(l) + delta^(l+1) * transpose(a^(l))
                        tri_W[l] += np.dot(delta[l+1][:,np.newaxis], np.transpose(a[l][:,np.newaxis]))# np.newaxis increase the number of dimensions
                        # trib^(l) = trib^(l) + delta^(l+1)
                        tri_b[l] += delta[l+1]
            # perform the gradient descent step for the weights in each layer
            for l in range(len(nn_structure) - 1, 0, -1):
                W[l] += -alpha * (1.0/N * tri_W[l])
                b[l] += -alpha * (1.0/N * tri_b[l])
            # complete the average cost calculation
            avg_cost = 1.0/N * avg_cost
            avg_cost_func.append(avg_cost)
            cnt += 1
        return W, b, avg_cost_func


def predict_y(W, b, X, n_layers):
    N = X.shape[0]
    y = np.zeros((N,))
    for i in range(N):
        a, z = feed_forward(X[i, :], W, b)
        y[i] = np.argmax(a[n_layers])
    return y
```

```python
nn_structure = [64, 30, 10]
    
# train the NN
W, b, avg_cost_func = train_nn(nn_structure, X_train, y_v_train, 3000)
```

```python
# get the prediction accuracy and print
y_pred = predict_y(W, b, X_test, 3)
print('Prediction accuracy is {}%'.format(accuracy_score(y_test, y_pred) * 100))
```

### Extended implementation using Numpy

```python
# Sigmoid activation function

def f(z):
    return 1 / (1 + np.exp(-z))

def f_deriv(z):
    return f(z) * (1 - f(z))

def softmax(z):
    return np.exp(z - np.max(z))/np.sum(np.exp(z - np.max(z)), axis=0, keepdims=True) 
```

```python
def feed_forward_with_softmax(x, W, b, nn_structure):
    a = {1: x} # create a dictionary for holding the a values for all levels
    z = { } # create a dictionary for holding the z values for all the layers
    for l in range(1, len(W) + 1): # for each layer
        node_in = a[l]
        z[l+1] = W[l].dot(node_in) + b[l]  # z^(l+1) = W^(l)*a^(l) + b^(l)
        
        # If the next layer is the output layer, use softmax
        if (z[l+1].shape[0] == nn_structure[-1]):
            a[l+1] = softmax(z[l+1])

        # Else, the next layer is a hidden layer, so use sigmoid 
        else: 
            a[l+1] = f(z[l+1]) # a^(l+1) = f(z^(l+1))

    return a, z
```

```python
def calculate_out_layer_delta(y, a_out, z_out):
    # delta^(nl) = -(y_i - a_i^(nl)) * f'(z_i^(nl))
    return -(y-a_out) * f_deriv(z_out) 

def calculate_out_layer_delta_with_softmax(y, a_out, z_out):
    return -(y-a_out)


def calculate_hidden_delta(delta_plus_1, w_l, z_l):
    return np.dot(np.transpose(w_l), delta_plus_1) * f_deriv(z_l)

```

```python
# Back propagation algorithm

def train_nn_with_softmax(nn_structure, X, y, iter_num=3000, alpha=0.25):
        W, b = setup_and_init_weights(nn_structure)
        cnt = 0
        N = len(y)
        avg_cost_func = []
        print('Starting gradient descent for {} iterations'.format(iter_num))
        while cnt < iter_num:
            if cnt%1000 == 0:
                print('Iteration {} of {}'.format(cnt, iter_num))
            tri_W, tri_b = init_tri_values(nn_structure)
            avg_cost = 0
            for i in range(N):
                delta = {}
                # perform the feed forward pass and return the stored a and z values, to be used in the
                # gradient descent step
                a, z = feed_forward_with_softmax(X[i, :], W, b, nn_structure)
                
                # loop from nl-1 to 1 backpropagating the errors
                for l in range(len(nn_structure), 0, -1):
                    if l == len(nn_structure):
                        delta[l] = calculate_out_layer_delta_with_softmax(y[i,:], a[l], z[l])
                        avg_cost += np.linalg.norm((y[i,:]-a[l]))
                    else:
                        if l > 1:
                            delta[l] = calculate_hidden_delta(delta[l+1], W[l], z[l])
 
                        tri_W[l] += np.dot(delta[l+1][:,np.newaxis], np.transpose(a[l][:,np.newaxis])) # np.newaxis increase the number of dimensions
                        # trib^(l) = trib^(l) + delta^(l+1)
                        tri_b[l] += delta[l+1]
            # perform the gradient descent step for the weights in each layer
            for l in range(len(nn_structure) - 1, 0, -1):
                W[l] += -alpha * (1.0/N * tri_W[l])
                b[l] += -alpha * (1.0/N * tri_b[l])
            # complete the average cost calculation
            avg_cost = 1.0/N * avg_cost
            avg_cost_func.append(avg_cost)
            cnt += 1
        return W, b, avg_cost_func


def predict_y(W, b, X, n_layers):
    N = X.shape[0]
    y = np.zeros((N,))
    for i in range(N):
        a, z = feed_forward(X[i, :], W, b)
        y[i] = np.argmax(a[n_layers])
    return y
```

```python
nn_structure = [64, 30, 10]
    
# train the NN
W, b, avg_cost_func = train_nn_with_softmax(nn_structure, X_train, y_v_train, 3000)
```

```python
# get the prediction accuracy and print
y_pred = predict_y(W, b, X_test, 3)
print('Prediction accuracy is {}%'.format(accuracy_score(y_test, y_pred) * 100))
```
